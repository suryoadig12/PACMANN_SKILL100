Contoh lain: Supermarket A memberikan harga diskon 15% kepada pelanggan yang bertransaksi belanja lebih dari Rp100.000. Wawan tertarik dengan promosi tersebut maka Wawan membeli kebutuhan Wawan dan ternyata didapatkan hasil:

Kondisi di atas dapat di dituliskan ke dalam pernyataan if, seperti berikut:

Kondisi di atas dapat di dituliskan ke dalam pernyataan `IF`, seperti berikut:

```python
total_belanja_wawan = 110_000
min_belanja = 100_000

# Kita buat sebuah kondisi jika total_belanja_wawan lebih besar atau sama dengan min_belanja diskon
if total_belanja_wawan >= min_belanja:
    print("mendapatkan diskon 15%")

    # Karena melewati kondisi, maka mendapatkan diskon 15%
    total_akhir = total_belanja_wawan - (total_belanja_wawan * 0.15)
    print(f"total belanja setelah diskon sebesar {total_akhir} rupiah")
```

Ternyata total belanja Wawan melebihi syarat ketentuan diskon, maka dia berhak mendapatkan diskon.

**Basic if/else**

**if/else** diperlukan ketika branching memiliki **2 kondisi**:

Misalkan, kamu pergi ke Pantai jika cuaca cerah, namun hari ini hujan maka kamu pergi ke Bioskop.

Maka ilustrasinya seperti berikut:

lsdjfaldsjf.JPG

Maka, jika ditulis bentuk pernyataan **if/else** menjadi:

```python
cuaca_hari_ini = "hujan"

# IF <condition> jika cuaca sebenarnya cerah sama dengan cerah
if cuaca_hari_ini == "cerah" :
    print("aku pergi kepantai")

# Else jika tidak memenuhi pernyataan if diatas
else:
    print("aku pergi ke bioskop")
```

Contoh lain: Wawan ingin membeli laptop dan harga laptop sebesar 20.000.000

```python
harga_laptop = 20_000_000
uang_wawan = 15_000_000

# Proses membandingkan uang wawan dan harga laptop
if uang_wawan >= harga_laptop:
    print("dapat membeli laptop")
else:
    print("tidak dapat membeli laptop")
```
