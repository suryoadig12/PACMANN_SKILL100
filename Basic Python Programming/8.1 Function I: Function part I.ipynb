Kita juga dapat memanfaatkan branching, looping, list, dan boolean.

```python
# Fungsi dengan mengimplementasikan brancing dan looping
# Jika produk merupakan produk promo, maka akan mendapat potongan harga 50%
# Jika bukan merupakan produk promo, maka tetap membayar harga normal

def diskon(promo, harga):
    if promo == True:
        potongan = 0.5
        total_harga = int(harga * potongan)
        print(total_harga)

    else:
        print(harga)
```

```python
diskon(True, 100_000)
```

```python
diskon(False, 100_000)
```

# **Default Value**



*   Pada contoh yang sudah kita bahas diatas, dapat disimpulkan bahwa setiap parameter pada fungsi harus memiliki sebuah nilai agar tidak terjadi error.

*   Selanjutnya, bagaimana dengan parameter yang sudah ditetapkan atau bersifat konstan. Apakah harus selalu didefinisikan ulang?
*   Untuk mengatasi hal ini, digunakan `default value` untuk memberikan nilai yang sudah tetap nilainya.


*   Dapat didefinisikan seperti berikut:



```python
# Fungsi dengan keyword arguments
def luas_lingkaran(jari_jari, PI = 3.14):
    print(f"Luas lingkaran dari jari-jari {jari_jari} adalah {PI*jari_jari*jari_jari}")

# Memanggil fungsi
luas_lingkaran(4)
luas_lingkaran(8)
luas_lingkaran(12)
```



*   Pada contoh diatas, phi merupakan nilai konstan bernilai 3.14.

*   Untuk mempersingkat pengerjaan program dan menghindari perulangan penulisan code, kita dapat menetapkan nilai parameter diawal atau bisa disebut dengan `default value`.
*   `default value` biasanya digunakan ketika kita tidak ingin repot mengisi semua parameter.


*   `default value` juga bisa digunakan ketika mengisi value parameter yang sudah ditetapkan.

# **Fungsi yang Mengembalikan Hasil (Returns)**

Dari segi pengembalian nilai, fungsi terbagi menjadi 2:

1. Fungsi yang tidak mengembalikan nilai
2. Fungsi yang mengembalikan nilai



*   Contoh-contoh sebelumnya merupakan fungsi yang tidak mengembalikan nilai.
*   Lalu apa perbedaan fungsi yang yang dapat mengembalikan nilai?

  *   Nilai yang dikembalikan suatu fungsi, bisa disimpan diolah kembali untuk berbagai kebutuhan.
  *   Penggunaan keyword `return` berfungsi untuk mengembalikan nilai.





asldjflajsld.JPG

```python
def luas_persegi_panjang (panjang, lebar):
    luas = panjang * lebar
    return luas
```

Nilai dari fungsi `luas_persegi_panjang` disimpan untuk kemudian dikembalikan nilainya.

```python
# Menghasilkan output

print('Luas persegi panjang dengan panjang 8 dan lebar 6:', luas_persegi_panjang(8, 6))
```



*   Sebuah pabrik karet memproduksi dan menjual karet keseluruh indonesia.

*   Kamu diminta untuk membuat **fungsi profit** yang menghitung keuntungan penjualan karet.

*   Rumus yang digunakan untuk membangun fungsi:

  *   hitung_profit = total_sales - total_cost
  *   total_sales = jumlah_produk * harga_jual

  *   total_cost = jumlah_produk * biaya_produksi


*   Menggunakan return untuk menyimpan nilai.


*   Kemudian, panggil fungsi yang dibuat!



```python
# Lengkapi titik-titik dibawah ini
# Membuat fungsi
def fungsi_profit(jumlah_produk, harga_jual, biaya_produksi):
  total_sales = jumlah_produk * harga_jual
  total_cost = jumlah_produk * biaya_produksi
  hitung_profit = total_sales - total_cost
  return hitung_profit

# Memanggil fungsi
fungsi_profit(jumlah_produk= 5, harga_jual= 20_000, biaya_produksi= 8_000)
```
