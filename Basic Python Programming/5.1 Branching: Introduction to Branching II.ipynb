**Else IF/Elif**

Elif diperlukan ketika branching memiliki lebih dari 2 kondisi.

Misalkan, kamu pergi ke Pantai jika cuaca cerah, jika cuaca hujan maka pergi ke Bioskop, jika tidak memenuhi cuaca keduanya kamu istirahat di Rumah, ternyata cuaca hari ini bersalju.

Maka ilustrasinya seperti berikut:

sdfoeuroasd.JPG

```python
cuaca_hari_ini = "Badai"

# IF kondisi jika cuaca cerah
if cuaca_hari_ini == "cerah":

    # Statement pertama
    print("Aku pergi ke Pantai")

# IF kondisi kedua jika cuaca hujan
elif cuaca_hari_ini == "hujan":

    # Statement kedua
    print("Aku pergi ke Bioskop")

# Else jika tidak memenuhi kedua kondisi sebelumnya
else:

    # Statement ketiga
    print("Aku istirahat di Rumah")
```

Sehingga, kondisi di atas jika di tuliskan ke dalam bentuk kode akan seperti berikut:

```python
# Inisialisasi variable
laptop_aser = 15_000_000
laptop_asas = 10_000_000
laptop_oman = 9_000_000
uang_wawan = 10_000_000

# Pembuatan branching
if uang_wawan >= laptop_aser:
    print("beli laptop 'aser'")
elif uang_wawan >= laptop_asas:
    print("beli laptop 'asas'")
elif uang_wawan >= laptop_oman:
    print("beli laptop 'oman'")
else:
    print("Menabung dulu sampai uang cukup!")
```

**Mini Exercise**

Anda makan di sebuah restoran. Anda ingin memesan menu sesuai dengan budget anda.



*   Jika uang anda lebih dari sama dengan 100.000 maka anda memakan steak.

*   Jika uang anda lebih dari sama dengan 50.000 maka anda memakan spaghetti.
*   Jika uang anda kurang dari 50.000 maka anda memakan nasi goreng.

Budget uang anda sebesar 60.000 maka lengkapilah program branching dibawah ini:

```python
# Lengkapilah program branching dibawah ini agar bisa makan spaghetti

budget = 60_000
if budget >= 100_000:
  print("Makan steak")
elif budget >= 50_000:
  print("Makan spaghetti")
else:
  print("Makan nasi goreng")
```

# **Branching + Boolean Logic**

Kita telah belajar branching dan juga boolean logic pada pertemuan sebelumnya, kita dapat mengkombinasikan branching dan boolean logic dalam bersamaan loh. Berikut contoh situasinya:

Dalam sebuah aplikasi e-learning memiliki dua tipe pengguna yang pertama adalah admin, dan kedua user. Dalam aplikasi ini memiliki 2 orang admin, admin pertama memiliki username bernama **RayhanNaufal**, yang kedua adalah **AriefRahman**, selain itu adalah user. Buatlah program yang mendeteksi apakah dia admin atau user.

```python
username = "RayhanNaufal"

if username == "RayhanNaufal" or username == "AriefRahman":
    print(f"Selamat datang Administrator {username}")
else:
    print(f"Selamat datang User {username}")
```
