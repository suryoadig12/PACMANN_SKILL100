Selain itu, kita juga dapat langsung mengakses fungsi dari modul dengan cara:

```python
# Mengimpor fungsi dalam modul
from nama_modul import fungsi

# Memanggil fungsi
fungsi()
```

Maka, jika ingin mengimpor fungsi **bunga_anuitas** dari modul **bunga_bank**

```python
# Mengakses modul dan fungsi
from bunga_bank import bunga_anuitas

# Memanggil fungsi
anuitas_1 = bunga_anuitas(2000000, 10)
print(f"Bunga Anuitas yang didapatkan adalah {anuitas_1}")
```

Jika mengimpor fungsi dengan cara diatas, maka pemanggilan fungsi ditulis dengan cara memanggil fungsi secara biasa.

Dalam mengakses modul, Fungsi juga dapat diimpor lebih dari satu atau beberapa fungsi tertentu yang diinginkan.

```python
# Mengakses modul dan fungsi
from bunga_bank import bunga_terendah, bunga_anuitas

# Memanggil fungsi
rendah_1 = bunga_terendah(2000000, 10)
anuitas_2 = bunga_anuitas(700000, 8)

print(f"Bunga terendah yang didapatkan adalah Rp {rendah_1}")
print(f"Bunga anuitas yang didapatkan adalah Rp {anuitas_2}")
```

Selain cara diatas, untuk mengakses semua fungsi yang ada pada modul, kita dapat menggunakan `*`.

```python
# Mengakses modul dan fungsi
from bunga_bank import *

rendah_1 = bunga_terendah(2000000, 10)
anuitas_2 = bunga_anuitas(700000, 8)

print(f"Bunga terendah yang didapatkan adalah Rp {rendah_1}")
print(f"Bunga anuitas yang didapatkan adalah Rp {anuitas_2}")
```



*   Perintah di atas akan mengimpor semua item yang ada pada modul **bunga_bank**.

*   Semua kecuali item dapat diimpor kecuali fungsi yang namanya diawali dengan tanda underscore `_`.
*   Contoh nama sebuah fungsi yang tidak dapat diimpor: fungsi `_bunga_harian()`.

**Mengganti Nama Modul dan Fungsi**



*   Jika nama modul atau nama fungsi terlalu panjang, atau ketika ingin merubah nama modul maupun fungsi agar lebih mudah ditulis ketika dipanggil.
*   Kita dapat menggunakan keyword `as` untuk mengubah nama modul maupun nama fungsi.



```python
# Mengubah nama modul

import nama_modul as nama_modul_baru
```

```python
# Mengubah nama fungsi

from nama_modul import nama_fungsi as nama_fungsi_baru
```

`as` berarti alias yang dapat mengubah nama modul menjadi sesuai keinginan user.

```python
# Contoh 1, mengubah nama modul

import bunga_bank as perhitungan_bunga
```



*   Pada contoh 1, **modul bunga_bank** berubah nama menjadi **perhitungan_bunga**
*   Pada contoh 2, fungsi **bunga_anuitas** berubah nama menjadi **anuitas**



Sehingga, jika memanggil fungsi dari modul, maka pemanggilan fungsi memakai nama baru dari fungsi yang telah ditetapkan.

```python
# Modul perhitungan_penjualan

def laba_bersih(pendapatan_kotor, harga_jual):
    pendapatan_bersih = pendapatan_kotor - harga_jual
    return pendapatan_bersih

def profit(jumlah_produk, harga_jual, biaya_produksi):
    total_penjualan = jumlah_produk * harga_jual
    total_biaya = jumlah_produk* biaya_produksi
    return total_penjualan - total_biaya
```



*   Potongan kode di atas adalah dari module **perhitungan_penjualan.py** dan ada di dalam folder `modul_video_learning_materi_8/`.

*   Dari modul **perhitungan_penjualan.py**, import modul tersebut dengan mengubah nama modul menjadi **penjualan**.
*   Import semua fungsi.


*   Panggil fungsi **profit** nilai input jumlah_produk = 150, harga_jual = 35000, biaya_produksi = 100000.



```python
# Import modul
from modul_video_learning_materi_8.perhitungan_penjualan as penjualan

# Panggil fungsi
penjualan.profit(150, 35000, 30000)
```
