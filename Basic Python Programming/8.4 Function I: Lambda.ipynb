# **Lambda Function**

**Apa itu Lambda Function?**



*   Di materi sebelumnya, kita sudah membahas tentang fungsi yang biasanya kita definisikan terlebih dahulu.

*   Dalam pemrograman Python, ada satu jenis fungsi lain yang dapat digunakan tanpa harus mendefinisikan nama fungsi tersebut (anonymous).
*   Fungsi spesial ini disebut `Lambda Function`.


*   Ketika fungsi biasa didefinisikan dengan keyword `def`, fungsi anonymous di definisikan dengan keyword `lambda`.



**Perbedaan Lambda Function Dengan Function Biasa**

alsdjflasdlf.JPG

Contoh sebuah fungsi biasa:

```python
# Fungsi biasa

def sentence(nama):
    sapa = f"Hello, {nama}"
    print(sapa)

sentence("World!")
sentence("Friends!")
```

Kemudian diubah menjadi format `lambda function`.

```python
# Fungsi lambda

sapa = lambda nama: print(f"Hello, {nama}")
sapa("World!")
sapa("Friends!")
```

Untuk mempersingkat kode program, kita bisa menggunakan fungsi lambda!

**Bagaimana Cara Membuat Lambda Function?**

**lambda** `<parameters>`: `<expression>`

Untuk membuat fungsi lambda dibutuhkan tiga bagian utama:



*   `lambda`: keyword untuk membuat fungsi lambda.

*   `parameters`: berisi parameter atau argumen fungsi
*   `expression`: isi fungsi.


Contoh:

Fungsi yang berisi perhitungan kuadrat

lsadjfeuorw.JPG

```python
# Sintaks penulisan fungsi lambda

kuadrat = lambda num: num * num
```



*   Karena fungsi lambda tidak punya nama, jadi kita butuh variabel untuk menyimpannya. Maka pada contoh diatas, variabel yang dipakai untuk menyimpan adalah **kuadrat**
*   Hampir sama seperti fungsi biasa, ketika ingin memanggil fungsi kita hanya perlu menuliskan nama variabelnya dan nilai input.



```python
# Memanggil fungsi lambda

kuadrat(5)
```



*   Seperti yang dilihat, bahwa fungsi `lambda` tidak bisa memiliki baris lebih dari satu.

*   Lambda juga tidak bisa memiliki lebih dari satu `expression`.
*   Meskipun begitu, fungsi lambda bisa memiliki `parameter` lebih dari satu.



Contoh:

Fungsi yang menghitung keuntungan penjualan produk.

```python
profit = lambda pendapatan, pengeluaran: pendapatan - pengeluaran

profit(1200000, 935000)
```



*   `pendapatan` dan `pengeluaran` adalah `parameter` yang berjumlah 2.

*   Hanya 1 `expression` yang hanya dapat dibuat dalam fungsi lambda, yaitu `pendapatan` - `pengeluaran`
*   Variabel `profit` dipanggil ketika ingin menjalankan fungsi lambda.


**Mini Exercise**



*   Sebuah supermarket berencana untuk mengadakan diskon akhir tahun untuk setiap jenis produknya.

*   Kamu diminta untuk membuat fungsi sederhana yang dapat menghitung harga produk yang harus dibayarkan setelah dipotong diskon.
*   Untuk testing program, ketika fungsi dipanggil nilai yang akan dimasukkan adalah:

  *   Jumlah produk = 2
  *   Harga produk = Rp. 1.000.000
  *   Diskon = 15%



```python
# Isilah titik-titik dibawah ini
# Membuat fungsi lambda

harga_diskon = lambda jumlah_produk, harga_produk, diskon: int(jumlah_produk*harga_produk) - ((jumlah_produk*harga_produk)*diskon)
```

Ketika fungsi dipanggil, fungsi akan menampilkan harga produk yang harus dibayarkan setelah diskon.

```python
# Memanggil fungsi lambda yang sudah dibuat

harga_diskon(2, 1_000_000, 0.15)
```

Fungsi Lambda juga dapat diimplementasikan untuk semua tipe data, percabangan, perulangan, dan boolean.

```python
#Fungsi biasa

def perhitungan(num):
    if num < 10:
        return num+1
    else:
        return 2*num
```

```python
perhitungan(10)
```

```python
#Kita ubah menjadi lambda

hitung = lambda num: num+1 if num < 10 else 2*num
```

```python
hitung(10)
```

**Mini Exercise**

Terdapat sebuah fungsi yang menentukan seorang karyawan boleh cuti atau tidak. Jika karyawan pernah lembur sebanyak lebih dari sama dengan delapan kali dan bolos kurang dari empat kali. Maka karyawan diperbolehkan untuk cuti.

```python
# Bentuk function biasa tanpa lambda

def cuti(lembur, bolos):
    if lembur >= 8 and bolos < 4 :
        print('Karyawan diperbolehkan cuti')
    else:
        print('Karyawan tidak diperbolehkan cuti')
```

```python
cuti(10, 3)
cuti(4, 5)
```

Mari kita ubah fungsi biasa kedalam bentuk lambda, kemudian panggil fungsi untuk mengecek apakah fungsi berjalan atau tidak.

```python
# Fungsi dalam bentuk lambda
# Lengkapi titik titik dibawah ini

cuti = lambda lembur, bolos: print("Karyawan diperbolehkan cuti") if lembur >= 8 and bolos < 4 else print("Karyawan tidak diperbolehkan cuti")
```

```python
# Memanggil fungsi lambda

cuti(10, 3)
cuti(4, 5)
```

**Mini Exercise**



*   Terdapat fungsi sederhana perhitungan margin laba kotor untuk kepentingan evaluasi penjualan suatu produk.
*   Agar program terlihat lebih sederhana, user ingin mengubah fungsi biasa kedalam bentuk lambda. Ubah `margin_kotor` kedalam fungsi `lambda`



```python
# Fungsi biasa
def margin_kotor(laba_kotor, pendapatan):
    hitung = (laba_kotor / pendapatan) * 100
    print(f"Margin laba kotor sebesar {int(hitung)}%")


# Memanggil fungsi
margin_kotor(150000, 250000)
```

```python
# Isilah titik-titik dibawah ini
# Fungsi lambda

hitung_margin_kotor = lambda laba_kotor, pendapatan: print(f"Margin laba kotor sebesar {int((laba_kotor/pendapatan)*100)}%")
```

```python
# Memanggil fungsi lambda

hitung_margin_kotor(150_000, 250_000)
```
